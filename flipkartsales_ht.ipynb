# Hypothesis testing

 H0:Discounted price makes an averge  of 2979.206104



 
 H1: Discounted price does not makes an averge  of 2979.206104

```python
samplec.discounted_price.mean()
```

### z test

```python
#standard deviation of the population
sd=np.std(samplec['discounted_price'])
sd
```

```python
samplec.std()
```

```python
xbar=samplec.discounted_price.mean()
mu=2979.20
n=1000 #sample size
alpha=0.05
```

```python
xbar
```

```python
zvalue=(xbar-mu)/(sd/np.sqrt(n))
zvalue
```

```python
z_critical = st.norm.ppf(alpha/2)  #One tail Test
z_critical
```

```python
if (zvalue<z_critical and abs(zvalue)> abs(z_critical)):
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## using p value

```python
p_val = (1 - st.norm.cdf(abs(zvalue))) * 2
p_val
```

```python
#comparing with alpha value
if p_val<alpha:
  print("Reject Null Hypotheisis")
else:
  print("Fail to reject Null Hypothesis")
```

## Hence discounted price on Average doesn't make 2979.20, i.e, we reject the Null Hypothesis

### T test

```python
tsamplec=sample.sample(28,replace=True)
tsamplec
```

```python
t_sd=np.std(tsamplec['discounted_price'])
t_sd
```

```python
xbar=tsamplec.discounted_price.mean()
print(xbar)
mu=2979.20
n=28 #sample size
alpha=0.05
```

```python
n = 28
degrees_of_freedom = n-1
t = (xbar - mu)/(t_sd / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha, degrees_of_freedom)
t_critical
```

```python
if (t > t_critical):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

### Both Z test and T test suggest the same result, The Null Hypothesis can be rejected.
